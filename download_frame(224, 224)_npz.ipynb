```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Folder input & output
input_dir = "/content/drive/MyDrive/satria_data_2025/df_train_videos"
output_dir = "/content/drive/MyDrive/Video Cache"
os.makedirs(output_dir, exist_ok=True)

# Ambil semua file mp4
video_files = [f for f in os.listdir(input_dir) if f.endswith(".mp4")]

# Loop dengan progress bar
for file in tqdm(video_files, desc="Processing videos", unit="video"):
    input_path = os.path.join(input_dir, file)
    output_path = os.path.join(output_dir, file.replace(".mp4", ".npz"))

    # Skip kalau file .npz sudah ada
    if os.path.exists(output_path):
        print(f"Skipped: {file} (already exists)")
        continue

    # Baca video dengan OpenCV
    cap = cv2.VideoCapture(input_path)
    fps = cap.get(cv2.CAP_PROP_FPS)  # FPS asli
    frame_interval = int(fps) if fps > 0 else 1  # ambil 1 frame per detik

    frames = []
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Simpan frame tiap interval tertentu
        if frame_count % frame_interval == 0:
            # Resize frame
            frame_resized = cv2.resize(frame, (224, 224))
            frames.append(frame_resized)

        frame_count += 1

    cap.release()

    # Simpan ke .npz
    frames_array = np.array(frames, dtype=np.uint8)
    np.savez_compressed(output_path, frames=frames_array)
    print(f"Saved: {output_path} ({len(frames)} frames)")
```
