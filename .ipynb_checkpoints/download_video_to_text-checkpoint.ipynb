```python
import os
from tqdm import tqdm

# Tambahkan kolom transcript kosong dulu
if "transcript" not in df_train_merged.columns:
    df_train_merged['transcript'] = ""

for idx, video_path in tqdm(df_train_merged['path'].items(), total=len(df_train_merged), desc="Extracting transcripts"):
    if os.path.exists(video_path):
        try:
            # Ambil teks maksimal 5 segmen × 60 detik = 300 detik
            text = extract_text_from_video(video_path, language='id-ID', max_duration=60, max_segments=5)
            text = preprocess_text(text) if text else ""
            df_train_merged.at[idx, 'transcript'] = text

        except Exception as e:
            print(f"[ERROR] Video index={idx}, path={video_path} gagal diproses: {e}")
            df_train_merged.at[idx, 'transcript'] = ""
    else:
        print(f"[WARNING] File tidak ditemukan (index={idx}): {video_path}")
        df_train_merged.at[idx, 'transcript'] = ""
```
